```python
def process_data_guarded(data):
    if not data:
        print("No data provided")
        return
    else:
        print(f"Processing data: {len(data)}")
        print("Data processed successfully")

process_data_guarded("Sample data")
process_data_guarded(None)
process_data_guarded([])
```

```python
import socket
def checkPort(host , port , timeout=5):
    '''
    checks if a tcp port is open on a given host

    :param host:
    :param port:
    :param timeout:
    :return:
    '''
    try:
        with socket.create_connection((host, port), timeout=timeout):
            return True
    except (socket.timeout, ConnectionRefusedError):
        return False

print (checkPort('localhost', 8888))
print(checkPort('localhost', 9999))
```

```python
def greet_users(names):
    '''Greet each user in the list of names.'''
    for name in names:
        print(f"Hello, {name.title()}!")

greet_users(['alice', 'bob', 'charlie'])
```
