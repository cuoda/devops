# Exercise: Log File Line Generator

You are a DevOps engineer tasked with creating a **memory-efficient** tool for parsing large log files. Loading an entire multi-gigabyte log file into memory is inefficient, so you need to process it **line by line** using a **generator**.

---

## Objective

Write a Python generator function:

```python
def read_log_lines(filepath):
    pass
```

The function should read a log file and yield only the **valid log entries**.

---

## Requirements

### Function Signature

The function must accept a single argument:

```python
filepath: str
```

which represents the path to the log file.

---

### Reading the File

- Open the file specified by `filepath`.
- Read the file **one line at a time** (do not load the entire file into memory).

---

### Valid Log Lines

A line is considered **valid** if:

- It is **not empty**.
- It is **not a comment**.

---

### Comment Lines

A comment line is any line that starts with `#`.

Examples:

```text
# This is a comment
```

```text
    # This is also a comment
```

Notice that leading whitespace before `#` should be ignored.

---

### Empty Lines

The following lines should be skipped:

```text

```

```text

```

(lines containing only whitespace)

---

### Whitespace Handling

Before yielding a valid line:

- Remove leading whitespace.
- Remove trailing whitespace.

For example:

```text
"   [INFO] Server started.   "
```

should become:

```text
"[INFO] Server started."
```

---

### Generator Requirement

Your solution **must** be a generator:

- Use the `yield` keyword.
- Do **not** return a list or any other collection.

---

# Example

## File: `sample.log`

```text
[INFO] Application starting...
[DEBUG] Connecting to database.

# Configuration section
  # A nested comment
[WARN] Deprecated feature used.

[ERROR] Failed to process request id: 123
```

---

## Expected Output

```text
[INFO] Application starting...
[DEBUG] Connecting to database.
[WARN] Deprecated feature used.
[ERROR] Failed to process request id: 123
```

---

# How Your Solution Will Be Tested

Your `read_log_lines()` function will be tested with:

1. A standard log file containing valid entries, comments, and empty lines.
2. An empty file.
3. A file containing only comments and blank lines.
4. A file where comment lines have leading whitespace.

---

# Goal

This exercise is designed to test:

- Generator functions (`yield`)
- File handling
- String processing (`strip()`, `startswith()`)
- Writing memory-efficient code for large files

````python
def read_log_lines(filepath):
    """
    Creates a generator that reads a log file, yielding valid, non-comment lines.

    Args:
        filepath (str): The path to the log file.

    Yields:
        str: A stripped, non-empty, non-comment line from the file.
    """
    with open(filepath, 'r') as file:
        for line in file:
            line = line.strip()
            if line and not line.startswith('#'):
                yield line



print(next(read_log_lines('sample.log')))
````
