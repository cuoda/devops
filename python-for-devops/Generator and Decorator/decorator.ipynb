````python
import time
def simple_task(sleep_duration):
    time.sleep(sleep_duration)
    print("Running a simple task")

def timing_decorator(original_func):
    def wrapper_func(*args, **kwargs):
        start = time.perf_counter()
        result = original_func(*args, **kwargs)
        end = time.perf_counter()
        print(f"Task {original_func.__name__!r} took {end - start:.4f} seconds")
        return result
    return wrapper_func

simple_task = timing_decorator(simple_task)
simple_task(2)
````

````python
@timing_decorator
def another_task(sleep_duration):
    time.sleep(sleep_duration)
    print("Running a simple task")

another_task(2)
````

````python
import  random

def retry(func):
    max_attempt = 5

    def wrapper(*args, **kwargs):
        for attempt in range(1, max_attempt+1):
            try:
                print(f"Attempt {attempt}/{max_attempt}...")
                return func(*args, **kwargs)
            except Exception as e:
                print(f"Attempt {attempt}/{max_attempt} failed: {e}")
                if attempt == max_attempt:
                    raise
        return None

    return wrapper

def sometimes_fails():
    if random.random() < 0.5:
        raise Exception("Random failure")
    print("Success!")

print(retry(sometimes_fails)())
````

````python
def log_calls_broken(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with args: {args}, kwargs: {kwargs}")
        value= func(*args, **kwargs)
        print(f"finished {func.__name__} with args: {args}, kwargs: {kwargs}")
        return value
    return wrapper

@log_calls_broken
def add(x, y):
    return x + y

print(f"Result: {add(2, 3)}")
````

````python
def log_and_raise(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"Error in {func.__name__} raised {e.__class__.__name__}")
            raise
    return wrapper

@log_and_raise
def divide(x, y):
    raise ValueError(" I cannot divide")

divide(3,5)
````

# log_and_raise Decorator

## Overview
`log_and_raise` is a Python decorator that logs exceptions raised by a function while preserving the original exception behavior.

## Function Signature
```python
def log_and_raise(func):
    """
    Decorator that catches exceptions, logs them, and re-raises them.

    Args:
        func: The function to be decorated

    Returns:
        wrapper: A wrapper function that logs and re-raises exceptions
    """
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            print(f"Error in {func.__name__} raised {e.__class__.__name__}")
            raise
    return wrapper

````python
from functools import wraps

def broken_decorator(func):
    @wraps(func)
    def wrapper (*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

@broken_decorator
def add(x, y):
    """
    Return sum of two numbers
    :param x:
    :param y:
    :return:
    """
    return x + y

print("Introspection without @wrgs:")
print(f"__name__ = {add.__name__}")
print(f"__doc__ = {add.__doc__}")
````

````python
from functools import wraps

def decorator_A(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with args: {args}, kwargs: {kwargs} from decorator_A")
        value = func(*args, **kwargs)
        print(f"Finished {func.__name__} with args: {args}, kwargs: {kwargs} of decorator_A")
        return value
    return wrapper

def decorator_B(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with args: {args}, kwargs: {kwargs} from decorator_B")
        value = func(*args, **kwargs)
        print(f"Finished {func.__name__} with args: {args}, kwargs: {kwargs} of decorator_B")
        return value
    return wrapper

@decorator_A
@decorator_B
def myFunc1():
    print("This is myFunc1")

@decorator_B
@decorator_A
def myFunc2():
    print("This is myFunc2")

myFunc1()
myFunc2()
````
