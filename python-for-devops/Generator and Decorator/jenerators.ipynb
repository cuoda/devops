````python
class CountToIterator:
     def __init__(self , max_count):
         self.max_count = max_count
         self.count = 1

     def __iter__(self):
           return self

     def __next__(self):
            if self.count > self.max_count:
                raise StopIteration
            else:
                current_count = self.count
                self.count += 1
                return current_count
````


## Custom Iterator Implementation

This section demonstrates how to create custom iterators in Python using the `__iter__()` and `__next__()` methods.

### Key Concepts:
- **`__iter__()`**: Makes an object iterable by returning an iterator object
- **`__next__()`**: Returns the next value from the iterator and raises `StopIteration` when done
- The pattern allows us to iterate over custom objects using `for` loops

### Example Use Case:
Creating a `CountTo` class that counts from 1 to a specified maximum value, demonstrating the iterator protocol.

````python
def count_up_to(limit):
    '''Generates numbers from 1 to the specified limit.
    Args:
        limit (int): The upper limit of the sequence.
    '''

    print("Generator function started")
    n = 1
    while n<= limit :
         print("yielding " ,n)
         yield n
         print(f"Resumed after yielding {n}.")
         n+=1

         print("Generator function ended")

count_gen = count_up_to(8)
print(f"count_gen is {count_gen} and type is {type(count_gen)}")
next(count_gen)

for count in count_gen:
    print(count)
````

## Generator: `count_up_to(limit)`

A simple generator function that yields numbers from 1 up to the specified `limit`.
The function also prints status messages to illustrate the generator's execution flow (what runs before the first `yield`, what runs when the generator is resumed, etc.).

Signature
- `count_up_to(limit: int) -> Generator[int, None, None]`

Description
- When the generator is created it does not execute the body until iteration begins.
- On the first `next()`/first iteration the function starts and prints `"Generator function started"`, then prints `"yielding <n>"` and yields the current number.
- After each `yield`, when iteration resumes, it prints `"Resumed after yielding <n>."`, increments `n`, and (as written) prints `"Generator function ended"` inside the loop (note: that line executes every loop iteration in the current implementation).
- When `n` becomes greater than `limit`, the generator finishes and Python raises `StopIteration`.

Parameters
- `limit` (int): the inclusive upper bound of the generated sequence (starts from 1).

Example usage
```python
# create generator
count_gen = count_up_to(3)
print(f"count_gen is {count_gen} and type is {type(count_gen)}")

# advance to first yield
next(count_gen)

# iterate through remaining values
for count in count_gen:
    print("Consumed value:", count)


````python
def filter_events(data):
    '''
    yeild only the even items from the input sequence
    Args:
    data (iterable(int or float)) : The data to iterate through and filter
    Returns:
    generator (int or float) : A generator object that yields the even items
    :param data:
    :return:
    '''

    print("filter_evens: starting")

    for item in data:
        if item % 2 == 0:
            print(f"filter_events: yielding {item}")
    print('filter_events: done')

filter_events(range(10))
````

## Generator: `demo_three_yields()`

A simple demonstration generator that yields three values and includes print statements to illustrate the generator lifecycle (when the function starts, what is yielded, what runs after each `yield`, and when the generator finishes).

Signature
- `demo_three_yields() -> Generator[Any, None, None]`

Description
- Calling `demo_three_yields()` returns a generator object; the function body does not execute until iteration begins.
- The generator yields three values (commonly 1, 2, 3 or three example payloads). Between yields there are print statements that demonstrate when execution resumes inside the generator.
- After the third yield the generator completes and a subsequent `next()` raises `StopIteration`.

Behavior notes
- The first call to `next()` runs the function until the first `yield`, printing any "start" or "yielding" messages encountered before yielding the first value.
- Each time the generator is resumed (by `next()` or continuing a `for` loop), the code after the previous `yield` runs until the next `yield` (or until function end).
- When the generator finishes, `StopIteration` is raised by `next()`; a `for` loop handles this automatically.

Examples

Create generator and advance manually:
```python
gen = demo_three_yields()
print(type(gen))        # generator object
first = next(gen)       # runs until first yield
print("got:", first)
second = next(gen)      # resumes and runs until second yield
print("got:", second)
third = next(gen)       # resumes and runs until third yield
print("got:", third)
# next(gen) would now raise StopIteration

````python
# normal function
def get_list_of_server():
    print("Regular function started.")
    servers = []

    for i in range(3):
        server_name=f"server{i+1}"
        servers.append(server_name)
        print(f"Regular function: {servers[i]}")

    print("Regular function ended.")
    return servers

servers = get_list_of_server()
print(servers)
````

````python
# generator function
def yield_server(count):
    print("generator function started.")


    for i in range(count):
        server_name=f"server{i+1}"
        print(f"\t Yielding {server_name}")
        yield server_name


    print("generator function ended.")

servers_gen = yield_server(5)

for server in servers_gen:
    print(f"generator function: {server}")
````

````python
import time

def load_data_eagerly(count , delay=0.1):
    '''
    Simulates loading 'count' items eagerly with a delay between each item.
    :param count:
    :param delay:
    :return:
    '''

    result =[]
    for i in range(count):
        time.sleep(delay)
        result.append(f"Data {i}")
    return result

t0 = time.time()
data = load_data_eagerly(5)
t1 = time.time()
print(f"regular : Returned {data} in {t1-t0:.2f} seconds")
````

````python
import time

def load_data_lazily(count , delay=0.1):
    '''
    Simulates loading 'count' items lazily with a delay between each item.
    :param count:
    :param delay:
    :return:
    '''


    for i in range(count):
        time.sleep(delay)
        yield f"Data {i}"

data_gen = load_data_lazily(5)
while True:
    t0 = time.time()
    try:
        item = next(data_gen)

    except StopIteration:
        break

    t1 = time.time()
    print(f"generator: Returned {item} in {t1-t0:.2f} seconds")
````
