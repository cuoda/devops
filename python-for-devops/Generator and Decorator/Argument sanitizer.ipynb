# Argument Sanitizer Decorator

## Requirements

* Create a decorator named `sanitize_hostname`.
* The wrapper function should accept arbitrary positional and keyword arguments (`*args`, `**kwargs`).
* Look for a keyword argument named `hostname`.
* If `hostname` is present:

  * Convert it to lowercase.
  * Remove leading and trailing whitespace.
* Call the original function with the original positional arguments and the updated `kwargs`.
* Return whatever value the original function returns.

## Function Skeleton

```python
def sanitize_hostname(func):
    def wrapper(*args, **kwargs):
        # sanitize hostname

        # call original function

        # return result
        pass

    return wrapper
```

## Example

```python
@sanitize_hostname
def connect_to_host(*, hostname):
    print(f"Connecting to sanitized hostname: '{hostname}'")
    return f"Connected to {hostname}"

connect_to_host(hostname="  PROD-API.local  ")
```

Expected sanitized value:

```python
"prod-api.local"
```


````python
import functools

def sanitize_hostname(func):
    """
    A decorator that finds a 'hostname' keyword argument, sanitizes it
    by converting it to lowercase and stripping whitespace, then passes it
    to the wrapped function.
    """
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        if 'hostname' in kwargs:
            kwargs['hostname'] = str(kwargs['hostname']).lower().strip()

        return func(*args, **kwargs)

    return wrapper


@sanitize_hostname
def connect_to_host(*, hostname):
    print(f"Connecting to sanitized hostname: '{hostname}'")
    return f"Connected to {hostname}"


connect_to_host(hostname="  PROD-API.local  ")
````
