```python
# 1. ingest the log lines
# 2. Filter log lines based on either level or message substring
# 3. Extract and return only the message attribute of the logs

import  sys
import  json

def read_log(filepath):
    """
    Reads the contents of a file line by line


    :param filepath: str - path to the file
    :return: generator (dict(str)) - the JSON dictionary of the log line
    """

    with open(filepath, 'r') as file:
       for line in file:
           line = line.strip()
           if not line:
               continue
           yield json.loads(line)

#function to filter logs
def filter_log(logs, level=None, message=None):
    """
    filters any iterable containing dictionaries based on the level and message attributes or both

    :param logs: (iterable(dict)) - iterable containing the logs to be filtered
    :param level: (str) - the level of the log to filter on
    :param message:  (str) - the message substring to filter on

    :return:
    generator (dict(str)) - the filtered logs
    """

    for log in logs:
        if level is not None and log.get('level' , "").lower() != level.lower():
            continue
        if message is not None and message.lower() not in log.get('message', '').lower():
            continue
        yield log

def extract_fields(logs , field="message"):
     """
     Extracts the specified field from the logs
     :param logs:  (iterable(dict)) - iterable containing the logs to be filtered
     :param field: (str) - the field to extract
     :return:
            generator (str) - the extracted field
     """
     for log in logs:
         yield log.get(field)

def get_first_n(logs , n=10):
    count = 0
    for log in logs:
        if count >= n:
            return
        yield log
        count += 1

logs_gen = read_log('large_logs.txt')
filter_gen = filter_log(logs_gen , message='user ')
extract_gen = extract_fields(filter_gen, 'message')

for log in get_first_n(extract_gen, 10):
    print(log)

print("generator object sizes (in bytes):",
      sys.getsizeof(logs_gen),
      sys.getsizeof(filter_gen),
      sys.getsizeof(extract_gen)
      )
```
