```python
import logging
import sys

for level in (
    logging.DEBUG,
    logging.INFO,
    logging.WARNING,
    logging.ERROR,
    logging.CRITICAL,
):
    print(f"{logging.getLevelName(level):8} = {level}")
```

```python
import sys
import logging

# Create logger
filter_logger = logging.getLogger("demo.filter")

# Logger level
filter_logger.setLevel(logging.INFO)

# Create handler
stream_handler = logging.StreamHandler(sys.stdout)

# Handler level
stream_handler.setLevel(logging.ERROR)

# Attach handler to logger
filter_logger.addHandler(stream_handler)

# Log messages
filter_logger.info("Info: will not be shown")
filter_logger.error("Error: will be shown")
```

```python
#configuring logs and handler

data_logger = logging.getLogger("demo.data")
data_logger.setLevel(logging.DEBUG)

data_sh = logging.StreamHandler(sys.stdout)
data_sh.setLevel(logging.ERROR)

data_fh = logging.FileHandler('process.log' , "w")
data_fh.setLevel(logging.INFO)

data_logger.addHandler(data_sh)
data_logger.addHandler(data_fh)

data_logger.debug("Debug: will not be shown")
data_logger.info("Info: will not be shown")
data_logger.warning("Warning: will be shown")
data_logger.error("Error: will be shown")
data_logger.critical("Critical: will be shown")
```
