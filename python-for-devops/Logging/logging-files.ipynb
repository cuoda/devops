```python
def cleanup_log_files(base_name:str):
    for file_names in os.listdir("."):
        if file_names.startswith(base_name):
            os.remove(file_names)
```

```python
import logging
import time

print("basic logging with filehandler")

basic_logger = logging.getLogger("file.basic")
basic_logger.setLevel(logging.DEBUG)

basic_fh = logging.FileHandler("basicFile.log" , delay=True)
basic_fh.setLevel(logging.INFO)

basic_logger.addHandler(basic_fh)

basic_logger.info("INFO: will be written in file")
```

```python
import logging
from logging.handlers import RotatingFileHandler
import os
import time
print("size-based log rotation with RotatingFileHandler")

rotating_logs_filename = "rotatingFile.log"

for file_name in os.listdir("."):
    if file_name.startswith(rotating_logs_filename):
        os.remove(file_name)

rotating_logger = logging.getLogger("file.rotating")
rotating_logger.setLevel(logging.DEBUG)

rotating_fh = RotatingFileHandler(
    rotating_logs_filename, maxBytes=1024, backupCount=2, encoding="utf-8"
)

rotating_fh.setFormatter(logging.Formatter("%(asctime)8s %(levelname)s %(message)s"))
rotating_logger.handlers.clear()
rotating_logger.addHandler(rotating_fh)

for i in range(30):
    rotating_logger.info(f"Entry {i}: {'Z' * 50}")
    time.sleep(0.05)
```

```python
import logging
from logging.handlers import TimedRotatingFileHandler
import os
import time

print("time-based log rotation with TimedRotatingFileHandler")

timed_rotating_logs_filename = "timedRotatingFile.log"
cleanup_log_files(timed_rotating_logs_filename)
timed_rotating_logger = logging.getLogger("file.timedRotating")
timed_rotating_logger.setLevel(logging.DEBUG)

timed_rotating_fh = TimedRotatingFileHandler(
    timed_rotating_logs_filename, when="S" , interval=3, backupCount=2, encoding="utf-8"
)

timed_rotating_fh.setFormatter(logging.Formatter("%(asctime)8s %(levelname)s %(message)s"))
timed_rotating_logger.handlers.clear()
timed_rotating_logger.addHandler(timed_rotating_fh)

for i in range(30):
    timed_rotating_logger.info(f"Entry {i}: {'Z' * 50}")
    time.sleep(0.05)
```
