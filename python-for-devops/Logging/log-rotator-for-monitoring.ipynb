# Log Rotator for a Monitoring Service

## Exercise: Log Rotator for a Monitoring Service

You are a DevOps engineer responsible for a long-running system monitoring service. This service continuously writes status updates to a log file. To prevent this file from consuming all available disk space, you need to implement a log rotation strategy based on file size.

Your task is to create a function `configure_rotating_logger` that sets up a logger to write to a file and automatically rotate it when it reaches a certain size.

---

## Functional Requirements:

The function must accept four arguments:

- `logger_name (str)`: The name for the logger.
- `log_filepath (str)`: The path to the log file.
- `max_size_bytes (int)`: The maximum size of the log file in bytes before it rotates.
- `backup_count (int)`: The number of old log files to keep.

It must get a logger instance and set its level to `DEBUG` to capture all messages.

It must create and configure a rotating file handler based on the provided arguments.

The handler should be added to the logger.

The function must return the fully configured logger object.

---

## Input validation:

- `logger_name` and `log_filepath` must be non-empty strings. Raise `ValueError` if empty, `TypeError` if not a string.

- `max_size_bytes` and `backup_count` must be integers. Raise `TypeError` if not.

- `max_size_bytes` must be greater than `0`. Raise `ValueError` if not.

- `backup_count` must be a non-negative integer (`0` or greater). Raise `ValueError` if negative.

---

## Example Usage:

```python
# Configure a logger that rotates after ~1KB and keeps 3 backups.
log_path = "monitor_service.log"

service_logger = configure_rotating_logger(
    'monitoring_service',
    log_path,
    max_size_bytes=1024,
    backup_count=3
)

# This will eventually cause the log file to rotate.
for i in range(200):
    service_logger.info(f"Checking status of component {i}.")
```

---

## How Your Solution Will Be Tested:

- The returned logger's level and handler configuration will be checked.

- The validation logic will be tested with invalid inputs for all arguments to ensure the correct `TypeError` or `ValueError` exceptions are raised.

- A test will write enough data to trigger a rotation to verify the behavior.

````python
import logging
from logging.handlers import RotatingFileHandler
import time
import os

def configure_rotating_logger(logger_name:str , log_filepath:str, max_size_bytes:int , backup_count:int):
    if not isinstance(logger_name, str) or not isinstance(log_filepath, str):
        raise TypeError("Logger name and log file path must be strings.")
    if not isinstance(max_size_bytes, int) or not isinstance(backup_count, int):
        raise TypeError("Max size and backup count must be integers.")
    if logger_name == "" or log_filepath == "":
        raise ValueError("Logger name and log file path cannot be empty strings.")
    if max_size_bytes <= 0 or backup_count < 0:
        raise ValueError(
    "max_size_bytes must be greater than 0 and backup_count must be non-negative."
)

    logger = logging.getLogger(logger_name)
    logger.setLevel(logging.DEBUG)

    handler = RotatingFileHandler(log_filepath, maxBytes=max_size_bytes, backupCount=backup_count)
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    handler.setFormatter(formatter)
    logger.handlers.clear()
    logger.addHandler(handler)

    return logger
````
