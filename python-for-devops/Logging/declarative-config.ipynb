```python
import logging
import logging.config

print("Declarative logging configuration- INI-file")

config_path="declarative-config.ini"

logging.config.fileConfig(fname=config_path, disable_existing_loggers=False)

app_logger = logging.getLogger("app")
app_logger.debug("INI-style message :- This is a debug message")
```

```python
import logging
import logging.config
from typing import Dict, Any
print("Declarative configuration using dictionary config")

dict_config: Dict[str, Any] = {
    "version": 1,
    "disable_existing_loggers": False,
    "formatters": {
        "simple":{
            "format": "%(asctime)s - %(name)s - %(levelname)s - %(message)s"
        }
    },
    "handlers": {
        "console": {
            "class": "logging.StreamHandler",
            "level": "INFO",
            "formatter": "simple",
            "stream": "ext://sys.stdout"
        }
    },
    "loggers": {
        "config.dict":{
            "level": "DEBUG",
            "handlers": ["console"]
        }
    }
}


logging.config.dictConfig(dict_config)
config_logger = logging.getLogger("config.dict")
config_logger.debug("Dictionary-style message :- This is a debug message")
config_logger.info("Dictionary-style message :- This is an info message")
```

```python
import json

print("Declarative configuration using json config")

with open("declarative-config.json", "r") as config_file:
    json_config = json.load(config_file)

logging.config.dictConfig(json_config)
config_logger = logging.getLogger("config.json")
config_logger.debug("JSON-style message :- This is a debug message")
config_logger.info("JSON-style message :- This is an info message")
```

```python
print("Dynamically building config")

base_config: Dict[str, Any] = {
    "version": 1,
    "disable_existing_loggers": False,
    "handlers": {},
    "formatters": {},
    "loggers": {}
}

base_config["formatters"]["simple"] = {
    "format": "%(asctime)s - %(name)s - %(levelname)s - %(message)s"
}

base_config["handlers"]["console"] = {
    "class": "logging.StreamHandler",
    "level": "DEBUG",
    "formatter": "simple",
    "stream": "ext://sys.stdout"
}

base_config["loggers"]["config.dynamic"] = {
    "level": "WARNING",
    "handlers": ["console"]
}

def is_debug():
    return True

if is_debug():
    base_config["loggers"]["config.dynamic"]["level"] = "DEBUG"

logging.config.dictConfig(base_config)
config_logger = logging.getLogger("config.dynamic")
config_logger.debug("Dynamic-style message :- This is a debug message")
config_logger.info("Dynamic-style message :- This is an info message")
```
