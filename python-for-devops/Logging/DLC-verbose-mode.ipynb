# Dynamic Logging Configuration with Verbose Mode

## Exercise: Dynamic Logging Configuration with Verbose Mode

You are building a command-line tool for your team. To help with troubleshooting, you want to add a `--verbose` flag that increases the tool's log output. When the flag is present, the tool should log detailed `DEBUG` messages; otherwise, it should only show `INFO` level messages and higher.

Your task is to implement a function, `configure_logging`, that sets up the application's logging based on a boolean flag. This function will start with a base configuration defined in a Python dictionary and modify it dynamically before applying it.

---

## Functional Requirements:

The function must accept one argument: `verbose` (a boolean).

Inside the function, you must define a base logging configuration as a dictionary. This dictionary should configure the root logger to use a `StreamHandler` and a simple `Formatter` with `%(levelname)s: %(message)s`. The default log level for the root logger should be `INFO`.

If the `verbose` argument is `True`, the function must programmatically modify the configuration dictionary to change the root logger's level to `DEBUG`.

If `verbose` is `False`, the configuration should be used as is (with the `INFO` level).

The function must return the root logger instance.

---

## Input Validation:

If the `verbose` argument is not a boolean (`True` or `False`), the function must raise a `TypeError`.

---

## How Your Solution Will Be Tested:

- When called with `verbose=False`, the logger should filter out `DEBUG` messages.

- When called with `verbose=True`, the logger should show `DEBUG` messages.

- The validation logic will be tested with non-boolean inputs to ensure a `TypeError` is raised.

```python
import logging
import logging.config

def configure_logging(verbose: bool):
    """
    Configures logging dynamically based on a verbose flag.

    Args:
        verbose (bool): If True, sets the root logger level to DEBUG.
                        Otherwise, the level remains INFO.

    Returns:
        logging.Logger: The configured root logger instance.

    Raises:
        TypeError: If verbose is not a boolean.
    """
    # TODO: Implement input validation.
    if not isinstance(verbose, bool):
        raise TypeError("verbose must be a boolean")

    # TODO: Implement base logging config dictionary.
    base_config = {
    "version": 1,
    "disable_existing_loggers": False,
    "formatters": {
        "simple": {
            "format": "%(levelname)s: %(message)s"
        }
    },
    "handlers": {
        "console": {
            "class": "logging.StreamHandler",
            "formatter": "simple",
            "stream": "ext://sys.stdout"
        }
    },
    "loggers": {
        "": {
            "level": "INFO",
            "handlers": ["console"]
        }
    }
}

    if verbose:
        base_config["loggers"][""]["level"] = "DEBUG"
        base_config["handlers"]["console"]["level"] = "DEBUG"

    # TODO: Apply the configuration to the logging library.
    logging.config.dictConfig(base_config)
    # TODO: Return the configured root logger instance.
    return logging.getLogger()
```
