# JSON Configuration Updater

**Exercise: JSON Configuration Updater**

You are a DevOps engineer responsible for automating application deployments. Part of your deployment pipeline involves updating a JSON configuration file to specify which version of a container image to deploy.

Your task is to implement a function, `update_image_tag`, that reads a JSON configuration file, updates the image tag for a specific service, and writes the changes back to the file.

---

## Functional Requirements:

The function must accept three arguments:

- `config_path` (`str` or a `pathlib.Path` object): The path to the JSON configuration file.

- `service_name` (`str`): The name of the service whose image tag needs to be updated.

- `new_tag` (`str`): The new image tag to set.

The function must update the `image_tag` key for the specified service with the `new_tag`.

After modification, the function must save the updated dictionary in the same JSON file.

The new JSON file must be written in a human-readable format with an indent of **4 spaces**.

The function should not return anything.

### Input Validation:

The function must validate its inputs:

- `config_path` must be a string or `Path` object. If not, raise `TypeError`. If it points to a non-existent file, raise `FileNotFoundError`.

- `service_name` and `new_tag` must be non-empty strings.

- Raise `TypeError` for incorrect types and `ValueError` for invalid values (for example, empty strings, non-existent file).

- If the specified `service_name` does not exist in the configuration, the function must raise a `KeyError`.

## How Your Solution Will Be Tested:

- A standard update scenario.

- A test to ensure the JSON output is correctly formatted with **4-space indents**.

- A test case where the `service_name` is not found, which must raise a `KeyError`.

- Validation tests for all function arguments.

````python
import json
from pathlib import Path

def update_image_tag(config_path: str | Path, service_name: str, new_tag: str) -> None:
    """
    Reads a JSON config file, updates a service's image tag, and writes it back.
    """
    # TODO: Implement input validation.
    # TODO: Read, update, and save updated JSON object.
    # Remember to use an indent of 4 for human-readable output.
    if not isinstance(config_path, (str, Path)):
        raise TypeError("config_path must be a string or Path object.")
    
    config_path = Path(config_path)

    if not config_path.is_file():
        raise FileNotFoundError("config_path must point to a valid file.")

    if not isinstance(service_name,str):
        raise TypeError("service_name must be a string.")

    if service_name == "":
        raise ValueError("service_name must be non-empty string.")

    if service_name not in config_path.read_text():
        raise KeyError(f"service_name {service_name} not found in config_path.")

    if not isinstance(new_tag,str):
        raise TypeError("new_tag must be a string.")

    if new_tag == "":
        raise ValueError("new_tag must be non-empty string.")

    config_dict = json.loads(config_path.read_text())
    config_dict["services"][service_name]["image_tag"] = new_tag
    config_path.write_text(json.dumps(config_dict, indent=4))

````
