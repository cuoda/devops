```python
import csv
from pathlib import Path

servers = Path("servers.csv")

try:
    with servers.open("r", encoding="utf-8", newline="") as f:
        reader = csv.reader(f)
        headers = next(reader)
        for idx, row in enumerate(reader, start=1):
            print(f"{idx}. {row}")
except FileNotFoundError:
    print("File not found.")
```
