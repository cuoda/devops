```python
import re


log_entry= "Ts=2023-10-27T12:00:00Z Level=ERROR User=admin Action=login_fail IP=18.0.0.1"

'''
our goal:-
1. group 1: the log level
2. group 2: the user name
3. group 3: the IP address
'''

pattern = r"Level=(\w+)\s+User=(\w+).*?\s+IP=([\d\.]+)"

match = re.search(pattern, log_entry)

if match:
    print(f"Full match:- {match.group(0)}")
    print(f"Level:- {match.group(1)}")
    print(f"user:- {match.group(2)}")
    print(f"IP:-  {match.group(3)}")
    print(f"All groups:- {match.group()}")
```

```python
import re

log_entry= "Ts=2023-10-27T12:00:00Z Level=ERROR User=admin Action=login_fail IP=18.0.0.1"

'''
our goal:-
1. group 1: the log level
2. group 2: the user name
3. group 3: the IP address
'''

pattern = r"Level=(?P<level>\w+)\s+User=(?P<user>\w+).*?\s+IP=(?P<ip>[\d\.]+)"

match = re.search(pattern, log_entry)

if match:
    print(f"Full match:- {match.group(0)}")
    print(f"Level:- {match.group('level')}")
    print(f"user:- {match.group('user')}")
    print(f"IP:-  {match.group('ip')}")
    print(f"All groups:- {match.group()}")
    print(f"group dictionary:- {match.groupdict()}")
```

```python
import re
log_line1 = "report.txt Status: OK"
log_line2 = "report Status: OK"
# Our goal:
#1. Group 1: The stem of the filename, with .txt being an optional string
#2. Group 2: The status code

pattern = r"^(.+?)(?:\.txt)?\s+Status:\s+(.+)$"

match_line1 = re.search(pattern, log_line1)
match_line2 = re.search(pattern, log_line2)

if match_line1: print(f"Line 1: {match_line1.group(1)}, {match_line1.group(2)}")
if match_line2: print(f"Line 2: {match_line2.group(1)}, {match_line2.group(2)}")
```

```python
import re

text = "This this is a test test."
pattern_numbers = r"(?i)\b(\w+)\s\1\b"
pattern_level = r"(?i)\b(?P<word>\w+)\s+(?P=word)\b"
print(f"Doubled words: {re.findall(pattern_numbers, text)}")
print(f"Level words: {re.findall(pattern_level, text)}")
```

```python
import re

text = "Error found: 404, 500, 403, 401. User IDs: user123,  admin99"
config = "timeout=60 retries=3 workers=5"

print(f"Numbers found: {re.findall(r"\d+?", text)}")
#find all in groups
print(f"key-value pairs found: {re.findall(r"(\w+)=(\w+)", config)}")

# finditer
for match in re.finditer(r"(\w+)=(\w+)", config):
    print(f"Key: {match.group(1)}, Value: {match.group(2)}")
```

```python
import re

text = "User IDs: user123, user456, user123456. Contact admin789 for help."

#Basic substitution
redacted = re.sub(r"user\d+", "[REDATED_USER]", text)
print(redacted)

#Back-reference for reusing information
partially_redacted = re.sub(r"(u)ser\d+(\d{2})", r"\1[REDATED_USER]\2", text)
print(partially_redacted)

#Limited count for substitutions
redacted_only_two = re.sub(r"(u)ser\d+(\d{2})", r"\1[REDATED_USER]\2", text, count=2)
print(redacted_only_two)

#Named groups for substitution
date_text = "Start: 2026-10-27 , End: 2027-01-23"
date_pattern_named= r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})"
replacement_format = r"\g<day>/\g<month>/\g<year>"
reformatted_date= re.sub(date_pattern_named, replacement_format, date_text)
print("Result date:-",reformatted_date)
```
