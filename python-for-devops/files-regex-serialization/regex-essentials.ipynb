```python
import re

line = "WARN: Disk usage at 91%"
pattern = r"WARN"
print(f"search '{pattern}' in line:", bool(re.search(pattern, line)))
print(f"match '{pattern}' in line:", bool(re.match(pattern, line)))
```

```python
import re
test = "Error code: E1234. cxge"
print(f"Dot find matches any character: {re.findall(r"c..e", test)}")
print(f"Start anchor ( does not finds 'E1234'): {re.findall(r"^E1234", test)}")
print(f"Start anchor (finds 'Error'): {re.findall(r"^Error", test, re.M)}")
print(f"End anchor (finds 'cxge'): {re.findall(r"cxge$", test)}")
print(f"Character set: {re.findall(r"[E0-9]+" , test)}")
```

```python
import re
my_income = "my Monthly income is $1000000.00 when i am 21 years old."

print(f"Find all numbers: {re.findall(r"\d+", my_income)}")
print(f"Find all string: {re.findall(r"\w+", my_income)}")
print(f"Whitespace: {re.findall(r"\s+", my_income)}")
print(f"Word boundary: {re.findall(r"\b\bincome\b", my_income)}")
```

```python
import re

text = 'aaa'
print(re.findall(r'a?' , text))
print(re.findall(r'a*', text))
print(re.findall(r'a+' , text))
print(re.findall(r'a{2}' , text))
print(re.findall(r'a{1,2}' , text))

print("Non-greedy matching")

print(re.findall(r'a*?' , text))
print(re.findall(r'a+?' , text))
print(re.findall(r'a{2}?' , text))
print(re.findall(r'a{1,2}?' , text))
```

```python
import re

html= "<p>One</p><p>Two</p><p>Three</p>"
print(f"Greedy: {re.findall(r"<.*>", html)}")
print(f"Non-greedy: {re.findall(r"<.??>", html)}")
```
