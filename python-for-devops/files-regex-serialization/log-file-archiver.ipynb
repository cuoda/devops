Your task is to implement a function, `archive_log_files`, that scans a given directory, finds all files with a `.log` extension, and renames them to include a specific date stamp.

## Functional Requirements

The function must accept two arguments:

- `log_directory` (a `str` or a `pathlib.Path` object): The directory to scan for log files. Your function should work correctly whether it receives a string path or a `Path` object.
- `archive_date` (a string): The date stamp to append to the filenames, in `YYYY-MM-DD` format.

```text
{original_stem}-{archive_date}.log
```
