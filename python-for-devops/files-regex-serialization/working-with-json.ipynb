```python
import json

from debugpy.launcher import output

api_response_str = '{"status": "active", "instance_id": "1-12345", "cores": 4, "tags": ["web", "prod"]}'

try:
    data = json.loads(api_response_str)
    print(data)
    print(type(data))
    print(f"Status: {data.get('status' , 'N/A')}")
    print(f"Instance ID: {data.get('instance_id' , 'N/A')}")
    print(f"Cores: {data.get('cores' , 'N/A')}")
    print(f"Tags: {data.get('tags' , 'N/A')}")
except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")
```

```python
import json
from pathlib import Path

config_path = Path("service.config.json")

with config_path.open("r") as f:
    config = json.load(f)

for service in config:
    service_name = service.get("service", None)
    if service_name:
        print(f"Service: {service_name}")
        print(f"Enabled: {service.get('enabled', False)}")
        print('-' * 20)
```
