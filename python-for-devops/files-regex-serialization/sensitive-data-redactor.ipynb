**Exercise: Sensitive Data Redactor in Configuration Files**

You are a DevOps engineer tasked with improving your company's security posture. Before archiving configuration files, all sensitive values like API keys and passwords must be redacted. The configuration files have a simple `key = value` or `key: value` format.

Your task is to implement a function, `redact_sensitive_data`, that reads a block of text, finds lines containing sensitive keys, and replaces their values with a placeholder.

---

## Functional Requirements:

The function must accept one argument: `content` (a string, which may be multi-line).

The function must identify lines that start with a sensitive key. The sensitive keys to look for are `api_key`, `password`, and `secret`.

The match must be case-insensitive (it should match `Password`, `API_KEY`, etc.).

The key can be followed by either an equals sign (`=`) or a colon (`:`), with optional whitespace around it.

The goal is to replace **only the value part of the line**, leaving the key and the separator intact. The value should be replaced with the placeholder `[REDACTED]`.

The function must return the redacted content as a single string.

### Input Validation:

If `content` is not a string, the function must raise a `TypeError`.

---

## Example Content:

```text

# Database Credentials

user = admin
```

---

## Example Usage:

```python
file_content = 'api_key = "abc-123"\nPassword: secret'
redacted_content = redact_sensitive_data(file_content)
print(redacted_content)
```
