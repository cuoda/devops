# User Login Event Parser

**Exercise: User Login Event Parser**

You are a security engineer for an internal application. To monitor user activity, you need to parse login event logs to identify who logged in and whether their attempt was successful.

Your task is to implement a function, `parse_login_event`, that uses a regular expression with named capturing groups to extract the username and status from a specific log format.

---

## Functional Requirements:

The function must accept one argument: `log_line` (a string).

It must parse log lines that follow the format: `LOGIN_EVENT: User '{username}' login attempt was {status}.`

The regular expression must use the following named capturing groups to retrieve the username and the login status respectively:

- **username:** To capture the username, which is enclosed in single quotes.

- **status:** To capture the final status word of the event (e.g., `SUCCESSFUL`, `FAILED`).

If the log line matches the pattern, the function should return a dictionary containing the data from these named groups.

If the log line does not match the pattern, the function should return `None`.

### Input validation:

If `log_line` is not a string, raise a `TypeError`.

---

## Example Usage:

```python
log1 = "LOGIN_EVENT: User 'jdoe' login attempt was SUCCESSFUL."
log2 = "LOGIN_EVENT: User 'admin' login attempt was FAILED."
log3 = "INFO: Application started."

print(parse_login_event(log1))  # Expected: {'username': 'jdoe', 'status': 'SUCCESSFUL'}
print(parse_log_event(log2))  # Expected: {'username': 'admin', 'status': 'FAILED'}
print(parse_login_event(log3))  # Expected: None
```

## How Your Solution Will Be Tested:

- A log line indicating a successful login.

- A log line indicating a failed login.

- A log line that does not match the login event format, which should return `None`.

- A test for a non-string input, which should raise a `TypeError`.

````python
import re
from typing import Optional

def parse_login_event(log_line: str) -> Optional[dict[str, str]]:
    """
    Parses a login event log line to extract the username and status.

    Args:
        log_line (str): The log line to parse.

    Returns:
        A dictionary with 'username' and 'status' if the line matches,
        otherwise None.

    Raises:
        TypeError: If log_line is not a string.
    """
    # TODO: Input validation

    # TODO: Define a regular expression pattern to match the log format:
    # "LOGIN_EVENT: User '{username}' login attempt was {status}."

    # TODO: If a match is found, return the dictionary of named groups; otherwise, return None.
    if not isinstance(log_line, str):
        raise TypeError(f"log_line must be a string")

    match = re.search(r"LOGIN_EVENT: User '(?P<username>[^']+)' login attempt was (?P<status>\w+).", log_line)
    if match:
        return match.groupdict()
    return None
````
