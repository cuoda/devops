# Log Line Error Detector

## Solution (Python)

```python
import re

def has_critical_error(log_line):
    """
    Returns True if the log line contains 'ERROR:' or 'FAIL:'
    (case-insensitive), otherwise returns False.
    """
    pattern = r"(ERROR:|FAIL:)"
    return bool(re.search(pattern, log_line, re.IGNORECASE))


# Example Usage
line1 = "2023-10-27 INFO: System started successfully."
line2 = "2023-10-27 ERROR: Database connection lost."
line3 = "2023-10-27 WARN: Disk usage high, but operation will not fail all clear."
line4 = "2023-10-27 DEBUG: User 'test' initiated a fail: sequence."

print(has_critical_error(line1))  # Expected: False
print(has_critical_error(line2))  # Expected: True
print(has_critical_error(line3))  # Expected: False
print(has_critical_error(line4))  # Expected: True
```

---

# Explanation

- `re.search()` searches the entire log line for a match.
- The regular expression:

```regex
(ERROR:|FAIL:)
```

matches either:
- `ERROR:`
- `FAIL:`

- The `re.IGNORECASE` flag makes the search case-insensitive, so it also matches:
  - `error:`
  - `Error:`
  - `FAIL:`
  - `fail:`
  - `FaIl:`
  - etc.

- `bool(re.search(...))` returns:
  - `True` if a match is found.
  - `False` otherwise.

---

# Functional Requirements

The function:

- Accepts one argument: `log_line` (a string).
- Uses a **single regular expression**.
- Checks for either the exact substring `ERROR:` or `FAIL:`.
- Performs a **case-insensitive** match.
- Returns `True` if either pattern exists anywhere in the string.
- Returns `False` if neither pattern is present.
