```python
try:
    with open("config.txt" , "+w" , encoding="utf-8") as f:
        f.write("config.txt\n")
        f.write("This is a new line")
    with open("config.txt" , "r" , encoding="utf-8") as f:
        print(">" , f.read() , "<")
except OSError as e:
    print(e)
```

```python
from pathlib import Path

path = Path("mode_demo.txt")

with path.open(mode="w" , encoding="utf-8") as f:
    f.write("This is a new line\n")

with path.open(mode="a" , encoding="utf-8") as f:
    f.write("appending a new line\n")

try:
    with path.open(mode="x" , encoding="utf-8") as f:
        f.write("This is a new line\n")

except FileExistsError as e:
    print("File already exists.")
```

```python
from pathlib import Path

sample = Path("sample.txt")
sample.write_text("This is a new line\n this is a 2nd line\n this is a 3rd line" , encoding="utf-8")

print("Iteration for reading the file")
with sample.open(mode="r", encoding="utf-8") as f:
    for line in f:
        print(line.strip())
```

```python
print("read() for reading the file")

with sample.open(mode="r", encoding="utf-8") as f:
    print(">" , f.read() , "<")
```

```python
print("readline() for reading the file")

with sample.open(mode="r", encoding="utf-8") as f:
    print(">" , f.readline(5) , "<")
```

```python
from pathlib import Path

write_demo = Path("write_demo.txt")

with write_demo.open(mode="w", encoding="utf-8") as f:
    f.write("This is a new line\n")
    f.write("This is a 2nd line\n")
    f.write("This is a 3rd line\n")

lines_to_write = [
    "user,ip,role\n",
    "admin,127.0.0.1,admin\n",
    "user,127.0.0.1,user\n"
]
with write_demo.open(mode="a", encoding="utf-8") as f:
    print(type(f"{line} " for line in lines_to_write))
    f.writelines(f"{line}" for line in lines_to_write)
```
