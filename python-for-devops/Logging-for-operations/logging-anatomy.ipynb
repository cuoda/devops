```python
import logging
from multiprocessing.process import name

root_logger = logging.getLogger()
print(f"Root Logger: name={root_logger.name}, level={root_logger.level}")

app_logger = logging.getLogger("app")
print(f"App Logger: name={app_logger.name}, level={app_logger.level}, parent={app_logger.parent.name}")

network_logger = logging.getLogger(('app.network'))
print(f"Network Logger: name={network_logger.name}, level={network_logger.level}, parent={network_logger.parent.name}")
```

```python
from logging import LogRecord

record = LogRecord(
    name="app.network",
    level=logging.INFO,
    pathname="app.py",
    lineno=10,
    msg="Connection established",
    args=(),
    exc_info=None
)

print("Log Record Contents:")

for attr in ("name", "levelname", "pathname", "lineno", "msg"):
    print(f"{attr}: {getattr(record, attr)}")
```

```python
import sys

demo_logger = logging.getLogger("handler_demo")
demo_logger.setLevel(logging.DEBUG)
demo_logger.handlers.clear()
print(f"Demo Logger: name={demo_logger.name}, level={demo_logger.level}")

stream_handler = logging.StreamHandler(sys.stdout)
stream_handler.setLevel(logging.WARNING)

demo_logger.addHandler(stream_handler)
demo_logger.debug("debug: This message should not be printed")
demo_logger.info("info: This message should not be printed")
demo_logger.warning("warning: This message should be printed")
demo_logger.error("error: This message should be printed")
demo_logger.critical("critical: This message should be printed")
```

```python
formatter = logging.Formatter(
    "%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)

stream_handler.setFormatter(formatter)
demo_logger.warning("This message should be printed with a timestamp and formatted")
```
