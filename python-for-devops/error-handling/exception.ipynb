```python
import inspect , builtins
from difflib import context_diff


def show_tree(base, level=0 , max_depth=1):
    if level > max_depth:
        return
    for name, obj in vars(builtins).items():
        if inspect.isclass(obj) and issubclass(obj, base) and obj is not base:
            print("\t" * level + f"- {name}")
            show_tree(obj , level+ 1 ,max_depth+ 1)

show_tree(Exception , max_depth=10)
```

## Explanation of `show_tree`

The `show_tree` function displays a hierarchy of built-in Python exception classes.

It starts from a given base class, such as `Exception`, and searches through Python’s built-in objects to find classes that inherit from that base class. Each matching subclass is printed with indentation, so the output looks like a tree structure.

### How it works

The function accepts three parameters:

- `base`: The class whose subclasses should be displayed.
- `level`: The current depth in the tree. It is used to control indentation.
- `max_depth`: The maximum depth up to which the tree should be printed.

First, the function checks whether the current level is greater than the allowed maximum depth. If it is, the function stops and returns.

Then, it loops through all objects available in Python’s built-in namespace. For each object, it checks whether:

1. The object is a class.
2. The object is a subclass of the current base class.
3. The object is not the same as the base class itself.

If all these conditions are true, the class name is printed. The indentation is created using tab characters, so subclasses appear nested under their parent classes.

After printing a subclass, the function calls itself again using that subclass as the new base. This recursive call allows the function to continue searching for deeper child classes.

### Purpose

The main purpose of `show_tree` is to visualize inheritance relationships between exception classes. For example, when called with `Exception`, it prints many built-in exception types such as `ArithmeticError`, `LookupError`, `RuntimeError`, and their subclasses.

This makes it easier to understand how Python exceptions are organized in an inheritance hierarchy.
