```python
def process_servers(server_list):
    if not isinstance(server_list, list):
        raise TypeError("input server_list must be a list")

    if len(server_list) == 0:
        raise ValueError("server_list must contain at least one server")

    print(f"Processing {len(server_list)} servers")

#process_servers([])
process_servers(["web01" , "web02"])
```

```python
def set_deployment_replicas(count):
    if not isinstance(count, int):
        raise TypeError(f"Replica count must be int , got {type(count).__name__}")

    if count <= 0 or count > 100:
        raise ValueError(f"Replica count must be between 1 and 100, got {count}")

    print(f"Setting deployment replicas to {count}")

for val in [5,-1, "three" , 150]:
    try:
        set_deployment_replicas(val)
    except (TypeError, ValueError) as e:
        print(f"Error: {e}")
```
