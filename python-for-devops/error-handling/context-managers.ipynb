```python
class MyContextManager:
    def __init__(self , timeout):
        self.timeout = timeout

    def __enter__(self):
        print("setup complete")
        return "a simple value"
    def __exit__(self , *args):
        print("cleanup complete")
        for arg in args:
            print(arg)
        return False

with MyContextManager(30) as cm:
    print(cm)
    print("Inside the block")
    raise ValueError("Error")
```

```python
import os
from contextlib import contextmanager

@contextmanager
def change_directory(destination):
    """
    Temporarily switch into destination. if the directory does not exist, create it.
    :param destination:  (str) path to the directory
    :return:
    """

    origin_dir = os.getcwd()

    try:
        print(f"Changing directory to {destination}")
        os.mkdir(destination)
        os.chdir(destination)
        yield os.getcwd()
    except FileExistsError as e:
        print(e)
        os.chdir(origin_dir)
        yield os.getcwd()
        print(f"Directory {destination} already exists")

    finally:
        print(f"Reverting to original dir: {origin_dir}")
        os.chdir(origin_dir)
print(f"start: {os.getcwd()}")

with change_directory("temp") as new_dir:
    print(f"new dir: {new_dir}")

print(f"end: {os.getcwd()}")
```
